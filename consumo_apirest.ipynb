```python
# imports
import requests
```

### Get todos estudantes

```python
api_url = "http://127.0.0.1:5000/estudantes"
response = requests.get(api_url)
```

```python
response.status_code
```

```python
response.json()
```

### Post estudante

```python
estudante = {"gender": "male", "race/ethnicity": "group B",
             "parental level of education": "bachelor's degree",
             "lunch": "standard", "test preparation course": "none",
             "math score" : 72, "reading score": 72, "writing score": 74}
api_url = "http://127.0.0.1:5000/estudantes"
response = requests.post(api_url, json=estudante)
```

```python
response.status_code
```

```python
response.json()
```

### PUT do estudante escolhido

```python
api_url = "http://127.0.0.1:5000/estudante/1000"
estudante = {"math score": 92, "race/ethnicity": "group A",
             "reading score": 92, "writing score": 94}
response = requests.put(api_url, json=estudante)
```

```python
response.status_code
```

```python
response.json()
```

### DELETE do estudante escolhido

```python
api_url = "http://127.0.0.1:5000/estudante/1000"
response = requests.delete(api_url)
```

```python
response.status_code
```

```python
response.json()
```
